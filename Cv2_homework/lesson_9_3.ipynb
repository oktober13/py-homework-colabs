```python
import cv2
```

```python
cap = cv2.VideoCapture('video/vid_cam.avi')
out = cv2.VideoWriter('video/output.avi', cv2.VideoWriter_fourcc(*'MJPG'), 10, (1280, 720))
cap.set(3, 1280)  #   
cap.set(4, 720)

ret, frame1 = cap.read()
ret, frame2 = cap.read()

while cap.isOpened():
    diff = cv2.absdiff(frame1, frame2)  #    ,        , ..         .
    gray = cv2.cvtColor(diff, cv2.COLOR_BGR2GRAY)  #    - 
    blur = cv2.GaussianBlur(gray, (5, 5), 0)  #   
    _, thresh = cv2.threshold(blur, 20, 255, cv2.THRESH_BINARY)  #       
    dilated = cv2.dilate(thresh, None, iterations=3)  #     erosion(), ..  ,       
    ontours, _ = cv2.findContours(dilated, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)  #    

    for contour in ontours:
        (x, y, w, h) = cv2.boundingRect(
            contour)  #          
        #  contourArea()   contour ,  ,        ,   
        # print(cv2.contourArea(contour))  - 
        if cv2.contourArea(contour) < 700:  #        700 px
            continue
        cv2.rectangle(frame1, (x, y), (x + w, y + h), (0, 255, 0), 2)  #     
        cv2.putText(frame1, "Status: {}".format("Dvigenie"), (20, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 1,
                    cv2.LINE_AA)  #  
    # cv2.drawContours(frame1, ontours, -1, (0, 255, 0), 2) -       
    out.write(frame1)
    cv2.imshow("frame1", frame1)
    frame1 = frame2
    ret, frame2 = cap.read()

    if cv2.waitKey(20) & 0xFF == ord('q'):  #     
        break

cap.release()
out.release()
cv2.destroyAllWindows()
```
