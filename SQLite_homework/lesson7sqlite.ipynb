### SQLite

Монтируем и создаем базу данных на нашем гугл диске

```python
import sqlite3 as sq

from google.colab import drive  # монтируем
drive.mount('/content/gdrive', force_remount=True)
con = sq.connect('/content/gdrive/My Drive/Colab Notebooks/sqlite_lesson/landmarks.db')  # создаем файл БД
cur = con.cursor()
```

Создаем таблицу в нашей базе

```python
cur.execute('''CREATE TABLE IF NOT EXISTS object (
  object_id INTEGER PRIMARY KEY AUTOINCREMENT,
  landmark TEXT NOT NULL,
  adress TEXT NOT NULL,
  price INTEGER NOT NULL DEFAULT 0
  )''')  # создали таблицу достопримечательностей
data = [(1, 'Храм Василия Блаженного', 'Москва, Красная площадь, 2', 500), (2, 'Кремль и Красная площадь', 'Москва, Тверской, Красная площадь, 1', 500), (3, 'Долина Гейзеров', 'Камчатский край, Елизовский район', 7000), (4, 'Эльбрус', 'Россия, гора Эльбрус', 21900), (5, 'Казанский Кремль', 'Россия, республика Татарстан, Казань, Кремлевская, 2', 0),
        (6, 'Нижегородский Кремль', 'Нижний Новгород, площадь Минина и Пожарского', 100), (7, 'Эрмитаж', 'Санкт-Петербург, Дворцовая площадь, 2', 400), (8, 'Петергоф', 'Петродворец, ул. Разводная, 2', 450)]  # данные для нашей таблицы
cur.executemany('''INSERT OR REPLACE INTO object VALUES(?, ?, ?, ?)''', data)
con.commit()
```

Посмотрим, что получилось

```python
cur.execute('''SELECT * FROM object''')
cur.fetchall()
```

## Посмотрим, куда можно сходить, если в кармане 500 рублей

```python
cur.execute('''SELECT * FROM object WHERE price <= 500''')
cur.fetchall()
```

## Расположим достопримечательности в порядке убывания цены
Чтобы порядок был убывающий, воспользуемся командой DESC

```python
cur.execute('''SELECT * FROM object ORDER BY price DESC''')
cur.fetchall()
```
