## Для построения графиков температуры, скорости ветра, относительной влажности, необходимо предварительно обработать и отфильтровать данные в этих столбцах

```python
time_date = pd.to_datetime(df['Местное время в Уфе / им. Мустая Карима (аэропорт)'][:700], format='%d.%m.%Y %H:%M')  # столбец даты и времени, тип дэйттайм
tempr = df['T'][:700].astype(float)  # столбец температуры, сразу указываем тип флоат
wind_speed = df['Ff'][:700].astype(int)  # столбец скорость ветра, тип инт
rel_humidity = df['U'][:700].astype(int)  # столбец относительная влажность, тип инт
```

# Построение графиков

## 1. График температуры

```python
plt.rcParams ['figure.figsize'] = [15, 10]  # увеличим окно графика
plt.plot(time_date, tempr, c='r', label='Изменение температуры')  # выведем график
plt.legend()
```

## График скорости ветра

```python
plt.plot(time_date, wind_speed, c='b', label='Изменение скорости ветра')
plt.legend()
```

## График относительной влажности

```python
plt.plot(time_date, rel_humidity, c='g', label='Изменение влажности')
plt.legend()
```

## Общий график

```python
plt.title("Метеографик для г. Уфа")
plt.xlabel("Дата")
plt.ylabel("Числовые данные соответственно - град. Цельсия, м/с, мм.рт.ст")
plt.plot(time_date, tempr, c='r', label='Изменение температуры, град. Цельсия')
plt.plot(time_date, wind_speed, c='b', label='Изменение скорости ветра, м/с')
plt.plot(time_date, rel_humidity, c='g', label='Изменение влажности, мм.рт.ст')
plt.legend()
```
